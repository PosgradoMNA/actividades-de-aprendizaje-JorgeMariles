- In **regression**, the labels are continuous quantities.
- Linear models predict by computing a weighted sum of input features plus a bias term.

## Simple Linear Regression

Simple linear regression equation:

$y = ax + b$
<br>
$a$: slope
<br>
$b$: intercept

Generate linear-looking data with the equation:

$y = 3X + 4 + noise$

```python
np.random.rand(100, 1)
```

```python
X = 2*np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
plt.scatter(X, y);
```

```python
import pandas as pd
pd.DataFrame(y)
```

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression(fit_intercept=True)
linear_reg.fit(X, y)
```

Plot the model's predictions:

```python
#X_fit[]
```

```python
# construct best fit line
X_fit = np.linspace(0, 2, 100)
y_fit = linear_reg.predict(X_fit[:, np.newaxis])

plt.scatter(X, y)
plt.plot(X_fit, y_fit, "r-", linewidth=2, label="Predictions")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14);
```

Predictions are a good fit.

Generate new data to make predictions with the model:

```python
y_new = linear_reg.predict(X_new)
y_new
```

## Polynomial Regression

If data is more complex than a straight line, you can use a linear model ti fit non-linear data adding powers of each feature as new features and then train a linear model on the extended set of features. 

$$y = a_0 + a_1x_1 + a_2x_2 + a_3x_3 + ...$$ to $$y = a_0 + a_1x + a_2x^2 + a_3x^3+ ...$$ 

This is still a linear model, the linearity refers to the fact that the coefficients never multiply or divide each other.

To generate polynomial data we use the function:

$y = 0.50X^2 + X + 2 + noise$

```python
# generate non-linear data e.g. quadratic equation
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10]);
```

```python
import pandas as pd
pd.DataFrame(y)
```

Now we can use `PolynomialFeatues` to transform training data adding the square of each feature as new features.

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
X_poly
```

`X_poly` now contains the original feature of X plus the square of the feature:

```python
print(X[0])
print(X[0]*X[0])
```

```python
X_poly[0]
```

Fit the model to this extended training data:

```python
lin_reg = LinearRegression(fit_intercept=True)
lin_reg.fit(X_poly, y)
lin_reg.coef_, lin_reg.intercept_
```

The model estimates:

$\hat{y} =  0.89X +0.48X^2 + 2.09$

Plot the data and the predictions:

```python
X_new=np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)
plt.plot(X, y, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-3, 3, 0, 10]);
```

## R square

R² es una medida estadística de qué tan cerca están los datos de la línea de regresión ajustada. También se conoce como el coeficiente de determinación o el coeficiente de determinación múltiple para la regresión múltiple. Para decirlo en un lenguaje más simple, R² es una medida de ajuste para los modelos de regresión lineal.



R² no indica si un modelo de regresión se ajusta adecuadamente a sus datos. Un buen modelo puede tener un valor R² bajo. Por otro lado, un modelo sesgado puede tener un valor alto de R².

 SSres + SSreg = SStot, R² = Explained variation / Total Variation

# Ejercicio 1

Utiliza la base de datos de 
https://www.kaggle.com/vinicius150987/manufacturing-cost

Suponga que trabaja como consultor de una empresa de nueva creación que busca desarrollar un modelo para estimar el costo de los bienes vendidos a medida que varían el volumen de producción (número de unidades producidas). La startup recopiló datos y le pidió que desarrollara un modelo para predecir su costo frente a la cantidad de unidades vendidas.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/marypazrf/bdd/main/EconomiesOfScale.csv')
df.sample(10)
```

```python
df.describe()
```

```python
X = df[['Number of Units']]
y = df['Manufacturing Cost']
```

```python
len(X)
```

```python
y.describe
```

```python
plt.plot(X,y,'b.')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 101)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
#lineal
lin_reg = LinearRegression(fit_intercept=True)
lin_reg.fit(X_train, y_train)
y_pred=lin_reg.predict(X_test)
print(F'ecuacion : X*{lin_reg.coef_[0]} + {lin_reg.intercept_}')
print('Error medio Absoluto (MAE):', metrics.mean_absolute_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('r2_score',r2_score(y_test, y_pred))

X_fit = np.linspace(1, 10, 100)
y_fit = lin_reg.predict(X_fit[:, np.newaxis])
plt.scatter(X, y)
plt.plot(X_fit, y_fit, "r-", linewidth=2, label="Predictions")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
```

```python
#polinomial
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X_train)


lin_reg = LinearRegression(fit_intercept=True)
lin_reg.fit(X_poly, y_train)
y_pred=lin_reg.predict(poly_features.fit_transform(X_test))
print(F'ecuacion : X^2*{lin_reg.coef_[0]}+X*{lin_reg.coef_[1]} + {lin_reg.intercept_}')
print('Error medio Absoluto (MAE):', metrics.mean_absolute_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('r2_score',r2_score(y_test, y_pred))

X_fit = np.linspace(1, 10, 100).reshape(-1, 1)
X_fit=poly_features.transform(X_fit)
y_fit = lin_reg.predict(X_fit)
plt.scatter(X, y)
plt.plot(X_fit[:,0], y_fit, "r-", linewidth=2, label="Predictions")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
```

```python
from sklearn.linear_model import Ridge
#ridge 
lin_reg = Ridge(fit_intercept=True)
lin_reg.fit(X_train, y_train)
y_pred=lin_reg.predict(X_test)
print(F'ecuacion : X*{lin_reg.coef_[0]} + {lin_reg.intercept_}')
print('Error medio Absoluto (MAE):', metrics.mean_absolute_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('r2_score',r2_score(y_test, y_pred))

X_fit = np.linspace(1, 10, 100)
y_fit = lin_reg.predict(X_fit[:, np.newaxis])
plt.scatter(X, y)
plt.plot(X_fit, y_fit, "r-", linewidth=2, label="Predictions")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
```

```python
from sklearn.linear_model import Lasso
#lasso 
lin_reg = Lasso(fit_intercept=True)
lin_reg.fit(X_train, y_train)
y_pred=lin_reg.predict(X_test)
print(F'ecuacion : X*{lin_reg.coef_[0]} + {lin_reg.intercept_}')
print('Error medio Absoluto (MAE):', metrics.mean_absolute_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('r2_score',r2_score(y_test, y_pred))

X_fit = np.linspace(1, 10, 100)
y_fit = lin_reg.predict(X_fit[:, np.newaxis])
plt.scatter(X, y)
plt.plot(X_fit, y_fit, "r-", linewidth=2, label="Predictions")
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
```

# Conclusiones 
El metodo que mas les conviene es el polinomial, se ajusta mucho mejor a la mayor distribucion de los datos apesar que en el extremo superior X el error pueda llegar a ser mas alto. Pero ya que la mayoria de las observaciones se encuentran en el medio de la distribucion no lo veo problematico, solo recomendaria discrecion en ambos extremos de los datos.
